```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
df = pd.read_csv('Salary_dataset.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Assumptions of Linear Regression
# 1. Linear Relationship b/w inputcolumn and output columns
# 2. No Multicorrelinity
# 3. Normal residual
# 4. Homoscasticity
# 5. no autocorrelinity residual
#
```

# 1. Linear Relationship b/w inputcolumn and output columns

```python
def scale(x):
    return x/100000
df['Salary'] = df['Salary'].apply(scale)
```

```python
df.head()
```

```python
plt.plot(df['YearsExperience'],df['Salary'],'o')
plt.xlabel('YearsExperience')
plt.show()
```

# 2. No Multicorrelinity

```python
# only single input c0lumn is available
```

# 3. Normal residual

```python
x_train,x_test,y_train,y_test = train_test_split(df.iloc[:,0:1],df.iloc[:,-1],test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
residual = y_pred - y_test
```

```python
sns.displot(residual,kind = 'kde')
```

# 4. Homoscasticity

```python
plt.scatter(residual,y_pred)
```

# 5. no autocorrelinity residual

```python
plt.scatter(residual)
```
