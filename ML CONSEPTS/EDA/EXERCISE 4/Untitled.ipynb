# According to Money column

```python
df.head()
```

```python
gr = df.groupby('Month_name').sum().reset_index()
gr
```

```python
month_order = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
               "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Convert 'mont' to an ordered categorical type
gr["Month_name"] =pd.Categorical(gr["Month_name"], categories=month_order, ordered=True)
gr = gr.sort_values("Month_name").reset_index(drop=True)
gr
```

```python
sns.lineplot(data = gr,x=gr['Month_name'],y=gr['money'])
# outcome:- two sesion comes for coffee sale in a year.
#          1. after march month there is a sharp decline in coffee sales because of weather change.
#          2. after july coffee sales starts to rise.
```

# pattern in weakday sales

```python
gr2 = df.groupby('Weekday').sum().reset_index()
gr2
```

```python
week_order = ['Sun','Mon','Tue','Wed','Thr','Fri','Sat']

# Convert 'mont' to an ordered categorical type
gr2["Weekday"] =pd.Categorical(gr2["Weekday"], categories=week_order, ordered=True)
gr2 = gr2.sort_values("Weekday").reset_index(drop=True)
gr2
```

```python
sns.lineplot(data = gr2,x=gr2['Weekday'],y=gr2['money'])
```
