## 3.2 Univariate Analysis For Categorical columns <span id='3.2'></span>


### 3.2.1 Which primary Device used most <span id='3.2.1'></span>


```python
df['Primary_Device'].value_counts().sort_values().plot(kind='bar')
```

##### <span style='color:deepskyblue'>Smartphone Primary device mostely used by indian kids</span>

### 3.2.2 Most of the child Suffers with <span id='3.2.2'></span>


```python
df['Health_Impacts'].value_counts().sort_values().plot(kind='bar')
```

##### <span style='color:deepskyblue'>Most of the Indian childs Suffering With Poor Sleep</span>

### 3.2.3 Most childs belongs to Urban Or Rural <span id='3.2.3'></span>


```python
df['Urban_or_Rural'].value_counts().sort_values().plot(kind='bar')
```

## 3.3 Gender wise daily screening time Distribution <span id='3.3'></span>


```python
sns.distplot(df[df['Gender']=='Male']['Avg_Daily_Screen_Time_hr'],hist=False,label='Male')
sns.distplot(df[df['Gender']=='Female']['Avg_Daily_Screen_Time_hr'],hist=False,label='Female')
plt.legend()
plt.show()
```

##### <span style='color:deepskyblue'>Screening time behaviour For both gender is Same</span>

## 3.4 Gender wise prefered Primary Device <span id='3.4'></span>


```python
Gender_wise_used_device=df.groupby('Gender')['Primary_Device'].value_counts().reset_index()
```

```python
sns.barplot(data=Gender_wise_used_device,y='count',x='Gender',hue='Primary_Device',palette='Blues')
```

## 3.5 Gender wise Exceeded recommended limit <span id='3.5'></span>


```python
sns.heatmap(pd.crosstab(df['Gender'],df['Exceeded_Recommended_Limit']))
```

##### <span style='color:deepskyblue'>This behaviour is also same for both gender.Slightely Mens are heigh in breaking recomended limit</span>

## 3.6 Gender wise Educational to recreational ratio <span id='3.6'></span>


```python
sns.distplot(df[df['Gender']=='Male']['Educational_to_Recreational_Ratio'],hist=False,label='Male')
sns.distplot(df[df['Gender']=='Female']['Educational_to_Recreational_Ratio'],hist=False,label='Female')
plt.legend()
plt.show()
```

##### <span style='color:deepskyblue'>Both geners having same behaviour</span>

## 3.7 Gender wise Health Impacts <span id='3.7'></span>


```python
df[df['Gender']=='Male']['Health_Impacts'].value_counts().sort_values().plot(kind='bar',label='Male',color='red')
df[df['Gender']=='Female']['Health_Impacts'].value_counts().sort_values().plot(kind='bar',label='Female')
plt.legend()
plt.show()
```

```python
df.groupby(['Health_Impacts', 'Gender']).size().unstack().plot(kind='bar', color=['red', 'blue'])

plt.legend(['Female', 'Male'])
plt.show()
```

## 3.8 Primary Device wise daily screening time <span id='3.8'></span>


```python
sns.kdeplot(data=df,x='Avg_Daily_Screen_Time_hr',hue='Primary_Device',fill=False)
plt.show()
```

## 3.9 Primary device wise health Impact <span id='3.9'></span>


```python
df.groupby(['Health_Impacts', 'Primary_Device']).size().unstack().plot(kind='bar', color=['red', 'blue','yellow','green'])

plt.legend()
plt.show()
```
