# 1. According to gender

```python
df.head(4)
```

```python
# are males spending more time on screen the womens
sns.distplot(df[df['Gender']== 'Male']["Avg_Daily_Screen_Time_hr"],hist=False,label='Male')
sns.distplot(df[df['Gender']== 'Female']['Avg_Daily_Screen_Time_hr'],hist = False,label='Female')
plt.legend(title='Category')
```

```python
sns.distplot(df[df['Gender']== 'Male']["Educational_to_Recreational_Ratio"],hist=False,label='Male')
sns.distplot(df[df['Gender']== 'Female']['Educational_to_Recreational_Ratio'],hist = False,label='Female')
plt.legend(title='Category')
```

```python
print('How much % Males Exceeded_Recommended_Limit ')
df[df['Gender']== 'Male']["Exceeded_Recommended_Limit"].value_counts().plot(kind='pie',autopct="%.2f")
```

```python
print('How much % Females Exceeded_Recommended_Limit ')
df[df['Gender']== 'Female']["Exceeded_Recommended_Limit"].value_counts().plot(kind='pie',autopct="%.2f")
```
