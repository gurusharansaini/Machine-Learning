# Exploratery Data Analysis

## Aggregate Expenditure trend of indian states

```python
df=ae
df_long = df.melt(id_vars=['State'], 
                  var_name='Year_Range', 
                  value_name='Expenditure')


df_long['Year'] = df_long['Year_Range'].str.split('-').str[0].astype(int)

df_long['Expenditure'] = pd.to_numeric(df_long['Expenditure'], errors='coerce')
```

```python
# मान लीजिए हम औसत खर्च (Average Expenditure) के आधार पर टॉप 5 राज्य चुनते हैं
top_states = df_long.groupby('State')['Expenditure'].mean().nlargest(5).index
df_filtered = df_long[df_long['State'].isin(top_states)]

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_filtered, x='Year', y='Expenditure', hue='State', marker='o', linewidth=2.5)
```
